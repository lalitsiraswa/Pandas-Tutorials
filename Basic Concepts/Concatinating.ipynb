```python
import numpy as np
import pandas as pd
```

```python
x1 = {'a' : [1, 1, 1, 1, 1], 'b' : [1, 1, 1, 1, 1], 'c' : [1, 1, 1, 1, 1], 'd' : [1, 1, 1, 1, 1], 'e' : [1, 1, 1, 1, 1]}
x2 = {'e' : [2, 2, 2, 2, 2], 'f' : [2, 2, 2, 2, 2], 'g' : [2, 2, 2, 2, 2], 'h' : [2, 2, 2, 2, 2], 'i' : [2, 2, 2, 2, 2]}
x3 = {'a' : [3, 3, 3, 3, 3], 'b' : [3, 3, 3, 3, 3], 'c' : [3, 3, 3, 3, 3], 'd' : [3, 3, 3, 3, 3], 'e' : [3, 3, 3, 3, 3]}
```

```python
df1 = pd.DataFrame(data = x1, index = [1, 2, 3, 4, 5])
df2 = pd.DataFrame(data = x2, index = [1, 2, 3, 4, 5])
df3 = pd.DataFrame(data = x3, index = [5, 6, 7, 8, 9])
print(df1, "\n")
print(df2, "\n")
print(df3, "\n")
```

```python
print(df1 + df2, "\n")
print(df1 + df3, "\n")
print(df2 + df3, "\n")
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df3], axis = 1)
```
